```python
import pandas as pd
from datetime import datetime
import random

rawpath = "../raw"
```

# Krakau

```python
df = pd.read_excel(
    f"{rawpath}/Krakau_Nachtrag.xlsx"
)
signaturen = df.einzelsignatur.to_list()
```

```python
def pica_schreiben(signatur):
    pica = f"""0500 Pa
0501 Text$btxt
0502 ohne Hilfsmittel zu benutzen$bn
0503 Blatt$bnb
0600 yy
1100 1939$n[zwischen 1939 und 1944]
1110 zwischen 1939 und 1944$a1939$b1944$4ezth
1130 TB-papier
1130 !040445224!
1131 !04046198X!
1132 a1-analog;a2-druck;f2-blatt
1500 /1pol/1ukr/1ger
1700 /1XA-DXDE
2105 04,P01
4000 Plakat Nr. WKII {signatur} der Plakatsammlung 1933-1945
4019 Plakat Nr. WKII {signatur} der Plakatsammlung 1933-1945
4030 [Distrikt Krakau] : [Verlag nicht ermittelbar]
4060 1 ungezähltes Blatt
4105 !118043062X!
4105 !1181223199!
4700 |BSM|
5590 [Objektgattung]
5590 !041133579!
7001 {now.strftime('%d-%m-%y')} : x
4800 !009033645!
6710 !1246327058!
6800 [Provenienz]
6800 !004709608! *1961-1993
6800 Stempel
7100 GS PL WKII @ m
7109 !!DBSM/S!! ; GS PL WKII {signatur}
8034 Dieses Plakat ist Teil eines Erschließungsprojekts und derzeit nicht ausleihbar.
\t\n"""
    return pica
```

```python
now = datetime.now()
# testfile
with open(f"../dat/wkII-krakau-{now.strftime('%y-%m-%d-%H-%M-%S')}.dat", "w") as f:
    with open("../dat/wkII-krakau-recent.dat", "w") as f2:
        for signatur in signaturen:
            f.write(pica_schreiben(signatur))
            f2.write(pica_schreiben(signatur))

with open(f"../dat/wkII-krakau-recent-sample.dat", "w") as f:
    for signatur in random.sample(signaturen, 5):
        f.write(pica_schreiben(signatur))
```
