```python
import pandas as pd
from datetime import datetime

rawpath = "../raw"
```

```python
df = pd.read_excel(
    f"{rawpath}/Wandzeitungen_WKII.xlsx"
)
df.rename({"sprache ":"sprache"}, axis=1, inplace=True)

df.sprachencode = df.sprachencode.str.strip()
df.sprache = df.sprache.str.strip()
df
```

```python
def pica_schreiben(row):
    pica = f"""0500 Pf
0501 Text$btxt
0501 unbewegtes Bild$bsti
0502 ohne Hilfsmittel zu benutzen$bn
0503 Blatt$bnb
0600 yy;lf
1100 {row['erj']}
1110 {row['erj']}$4ezth
1131 !04046198X!
1500 {row['sprachencode']}
2105 04,P01
4000 {row['idn abvz']}
4004 *{row['bandnr']}*
4060 1 ungez√§hltes Blatt
4061 Illustrationen
4105 !118043062X!
4201 Plakatseriennummer: {row['anmerkung']}
4221 Text {row['sprache']}
4700 |BSM|"""
    if type(row['interne anmerkung']) != float:
        pica += f" *{row['interne anmerkung']}"

    pica += f"""
5590 [Objektgattung]
5590 !041133579!
6710 !1059794659!
7001 {now.strftime('%d-%m-%y')} : x
4800 !009033645!
"""
    if type(row['zustand']) != float:
        pica += f"4801 {row['zustand']}\n"
    
    pica += f"""7100 GS PL WKII @ m
7109 !!DBSM/S!! ; GS PL {row['einzelsignatur']}
\t\n"""
    return pica
```

```python
now = datetime.now()
# testfile
with open(f"../dat/wkII-wandzeitungen-{now.strftime('%y-%m-%d-%H-%M-%S')}.dat", "w") as f:
    with open("../dat/wkII-wandzeitungen-recent.dat", "w") as f2:
        for index, row in df.iterrows():
            f.write(pica_schreiben(row))
            f2.write(pica_schreiben(row))

with open(f"../dat/wkII-wandzeitungen-recent-sample.dat", "w") as f:
    for index, row in df.sample(5).iterrows():
        f.write(pica_schreiben(row))
```
